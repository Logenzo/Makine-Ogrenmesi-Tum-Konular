# **LendingClub veri setine uygulanan Random Forest**

Bu alıştırma için LendingClub.com'dan kamuya açık verileri inceleyeceğiz. Lending Club, paraya ihtiyacı olan kişileri (borç alanlar) parası olan kişilerle (yatırımcılar) buluşturuyor. Kredi ile ilgili çok çeşitli veriler göz önüne alındığında birine borç para verme riskini tahmin etmek için bir model oluşturmaya çalışacağız. 2007-2010 yılları arasındaki borç verme verilerini kullanacağız ve borç alan kişinin kredisini tam olarak geri ödeyip ödemediğini sınıflandırmaya ve tahmin etmeye çalışacağız.

İşte veri setindeki sütunların temsil ettikleri:

▶ **credit.policy**: Müşteri LendingClub.com'un kredi değerlendirme kriterlerini karşılıyorsa 1, karşılamıyorsa 0.

▶ **purpose**: Kredinin amacı ("credit_card", "debt_consolidation", "educational", "major_purchase", "small_business" ve "all_other" değerlerini alır).

▶ **int.rate**: Kredinin faiz oranı, bir oran olarak (%11'lik bir oran 0,11 olarak saklanacaktır). LendingClub.com tarafından daha riskli olarak değerlendirilen borçlulara daha yüksek faiz oranları atanır.

▶ installment: Kredinin finanse edilmesi durumunda borçlunun borçlu olduğu aylık taksitler.

▶ **log.annual.inc**: Borçlunun kendi bildirdiği yıllık gelirinin doğal logu.

▶ **dti**: Borçlunun borç-gelir oranı (borç miktarının yıllık gelire bölünmesi)

▶ **fico**: Borçlunun FICO kredi puanı.

▶ **day.with.cr.line**: Borçlunun kredi limitine sahip olduğu gün sayısı.

▶ **revol.bal**: Borçlunun döner bakiyesi (kredi kartı fatura döngüsünün sonunda ödenmemiş tutar)

▶ **revol.util**: Borçlunun döner hat kullanım oranı (mevcut toplam krediye göre kullanılan kredi limiti miktarı)

▶ **inq.last.6mths**: Borçlunun son 6 ay içinde alacaklılar tarafından yapılan sorgulama sayısı.

▶ **delinq.2yrs**: Borçlunun son 2 yıl içinde bir ödemeyi 30+ gün geciktirme sayısı.

▶ **pub.rec**: Borçlunun kamuya açık kayıtlarının sayısı (iflas başvuruları, vergi hacizleri veya yargı kararları)

▶ **not.fully.paid**: Sınıflandırma için faiz miktarı - borçlunun parayı tam olarak geri ödeyip ödemediği

# **Verilerin Ayarlanması**

***Kategorik Özellikler***

Amaç sütunu kategoriktir. Bunları kukla değişkenler kullanarak dönüştürüyoruz, böylece sklearn bunları anlayabilecek.

```python
df_final = pd.get_dummies(df,['purpose'],drop_first=True)
```

```python
df_final.head()
```

# **Train Test Bölünmesi**



```python
from sklearn.model_selection import train_test_split
X = df_final.drop('not.fully.paid',axis=1)
y = df_final['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
X.head()
```

# **Bir Karar Ağacı Modelinin Eğitilmesi**



```python
from sklearn.tree import DecisionTreeClassifier
```

***dtree adında bir DecisionTreeClassifier() örneği oluşturun ve bunu eğitim verilerine uydurun.***

```python
dtree = DecisionTreeClassifier(criterion='gini', max_depth=None)
```

```python
dtree.fit(X_train,y_train)
```

# **Karar Ağacının Tahminleri ve Değerlendirilmesi**

***Test setinden tahminler oluşturun ve bir sınıflandırma raporu ve bir karışıklık matrisi oluşturun.***

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
cm=confusion_matrix(y_test,predictions)
print(cm)
print("Tahmin doğruluğu:",round((cm[0,0]+cm[1,1])/cm.sum(),3))
```

# **Random Forest Modelinin Eğitimi**

Şimdi modelimizi eğitme zamanı!

RandomForestClassifier sınıfının bir örneğini oluşturun ve önceki adımdaki eğitim verilerimize uydurun.



```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train, y_train)
```

# **Tahminler ve Değerlendirme**

y_test değerlerini tahmin edelim ve modelimizi değerlendirelim.

X_test verileri için not.fully.paid sınıfını tahmin edin

```python
rfc_pred = rfc.predict(X_test)
```

*Şimdi sonuçlardan bir sınıflandırma raporu oluşturun. Garip bir şey veya bir tür uyarı alıyor musunuz ?*

```python
cr = classification_report(y_test,predictions)
```

```python
print(cr)
```

***Tahminler için Karışıklık Matrisini gösterin.***

```python
cm = confusion_matrix(y_test,rfc_pred)
print(cm)
```

***Rastgele ormanda artan ağaç sayısıyla bir döngü çalıştırmak ve karışıklık matrisinin doğruluğunu kontrol etmek***

**'Gini' veya 'entropi' kriteri**

```python
nsimu = 21
accuracy=[0]*nsimu
ntree = [0]*nsimu
for i in range(1,nsimu):
  rfc = RandomForestClassifier(n_estimators=i*5,min_samples_split=10,max_depth=None,criterion='gini')
  rfc.fit(X_train, y_train)
  rfc_pred = rfc.predict(X_test)
  cm = confusion_matrix(y_test,rfc_pred)
  accuracy[i] = (cm[0,0]+cm[1,1])/cm.sum()
  ntree[i]=i*5
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=ntree[1:nsimu],y=accuracy[1:nsimu],s=60,c='red')
plt.title("Number of trees in the Random Forest vs. prediction accuracy (criterion: 'gini')", fontsize=18)
plt.xlabel("Number of trees", fontsize=15)
plt.ylabel("Prediction accuracy from confusion matrix", fontsize=15)
```

```python
nsimu = 21
accuracy=[0]*nsimu
ntree = [0]*nsimu
for i in range(1,nsimu):
  rfc = RandomForestClassifier(n_estimators=i*5,min_samples_split=10,max_depth=None,criterion='entropy')
  rfc.fit(X_train, y_train)
  rfc_pred = rfc.predict(X_test)
  cm = confusion_matrix(y_test,rfc_pred)
  accuracy[i] = (cm[0,0]+cm[1,1]/cm.sum())
  ntree[i]=i*5
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=ntree[1:nsimu],y=accuracy[1:nsimu],s=60,c='red')
plt.title("Number of trees in the Random Forest vs. prediction accuracy (criterion: 'entropy')", fontsize=18)
plt.xlabel("Number of trees", fontsize=15)
plt.ylabel("Prediction accuracy form confusion matrix", fontsize=15)
```
