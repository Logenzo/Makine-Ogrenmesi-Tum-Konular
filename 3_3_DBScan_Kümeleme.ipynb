# **DBSCAN Kümeleme**

DBSCAN algoritması kümeleri düşük yoğunluklu alanlarla ayrılmış yüksek yoğunluklu alanlar olarak görür.

***Bu oldukça genel görüş nedeniyle, kümelerin dışbükey şekilli olduğunu varsayan k-ortalamanın aksine, DBSCAN tarafından bulunan kümeler herhangi bir şekilde olabilir.***

***DBSCAN'in merkezi bileşeni, yüksek yoğunluklu alanlarda bulunan örnekler olan çekirdek örnekler kavramıdır. Bu nedenle bir küme, her biri birbirine yakın olan (bazı mesafe ölçüleriyle ölçülen) bir dizi çekirdek örnek ve bir çekirdek örneğe yakın olan (ancak kendileri çekirdek örnek olmayan) bir dizi çekirdek olmayan örnekten oluşur.***

***Kütüphaneleri içe aktarın***

```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Ay şeklinde ve çörek şeklinde veriler oluşturun**

```python
from sklearn import cluster, datasets
```

```python
n_samples = 1500
noisy_circles, circle_labels = datasets.make_circles(n_samples = n_samples, factor = .5,
                                                     noise = .05)
noisy_moons, moon_labels = datasets.make_moons(n_samples = n_samples, noise = .1)
```

```python
noisy_moons = np.array(noisy_moons)
noisy_circles = np.array(noisy_circles)
```

```python
plt.figure(figsize=(8,5))
plt.title("Half-moon shaped data", fontsize=18)
plt.grid(True)
plt.scatter(noisy_moons[:, 0], noisy_moons[:, 1])
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title("Concentric circles of data points", fontsize=18)
plt.grid(True)
plt.scatter(noisy_circles[:, 0], noisy_circles[:, 1])
plt.show()
```

# **k-ortalamalar doğru kümeleri belirleyebilir mi?**

```python
km = cluster.KMeans(n_clusters = 2)
```

```python
km.fit(noisy_moons)
```

```python
plt.figure(figsize=(8,5))
plt.title("Half-moon shape data", fontsize=18)
plt.grid(True)
plt.scatter(noisy_moons[:, 0],noisy_moons[:, 1], c=km.labels_)
plt.show()
```

```python
km.fit(noisy_circles)
```

```python
 plt.figure(figsize=(8,5))
 plt.title("Concentric circles of data points", fontsize=18)
 plt.grid(True)
 plt.scatter(noisy_circles[:, 0], noisy_circles[:, 1], c=km.labels_)
 plt.show()
```

# **DBSCAN nasıl performans gösteriyor?**

```python
dbs = cluster.DBSCAN(eps=0.1)
```

```python
dbs.fit(noisy_moons)
```

```python
plt.figure(figsize=(8,5))
plt.title("Half-moon shaped data", fontsize=18)
plt.grid(True)
plt.scatter(noisy_moons[:, 0], noisy_moons[:, 1], c=dbs.labels_)
plt.show()
```

```python
dbs.fit(noisy_circles)
```

```python
plt.figure(figsize=(8,5))
plt.title("Concentric circles of data points", fontsize=18)
plt.grid(True)
plt.scatter(noisy_circles[:, 0], noisy_circles[:, 1], c=dbs.labels_)
plt.show()
```
